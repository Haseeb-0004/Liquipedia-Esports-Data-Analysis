***Opening And ELimination Stages***

```python
def extract_link_text(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    h1_tag = soup.find('h1', {'class': 'firstHeading'})
    if h1_tag:
        return h1_tag.get_text(strip=True)
    return None

url = "https://liquipedia.net/counterstrike/PGL/2021/Stockholm"
response = requests.get(url)

if response.status_code == 200:     #if greater than 200 cannot extract
    soup = BeautifulSoup(response.content, "html.parser")
    
    tables = soup.find_all("table", class_="wikitable wikitable-bordered wikitable-striped swisstable")
    # tables = soup.find_all("table", class_="table table-bordered table-striped swisstable")#inspect element second table

    count = 0
    for table in tables:
        headers = [header.text.strip() for header in table.find('tr')]
        
        data = []
        rows = table.find_all("tr")

        for row in rows:
            cells = []
            for cell in row.find_all(['td', 'th']):
                link = cell.find('a')
                if link:
                    url = link.get('href')
                    if url:
                        if not url.startswith('http'):
                            url = 'https://liquipedia.net' + url
                        link_text = extract_link_text(url)
                        cells.append(link_text)
                else:
                    cells.append(cell.get_text(strip=True))
            
            if cells:  
                data.append(cells)
        
        data.pop(0) # Removing headers
        
        if data:
            df = pd.DataFrame(data, columns=headers)
            df.drop(['BU','#'], axis=1, inplace=True)
            if count == 0:
                df_o = df
            else:
                df_e = df
            
            count += 1
# df_e
```
