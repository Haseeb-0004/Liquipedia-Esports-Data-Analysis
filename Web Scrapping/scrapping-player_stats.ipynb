```python
import requests
from bs4 import BeautifulSoup

url = "https://bo3.gg/matches/faze-vs-natus-vincere-31-03-2024"

response = requests.get(url)

if response.status_code == 200:

    soup = BeautifulSoup(response.content, "html.parser")

    rows = soup.find_all('div', class_='table-row')

    print("Player Stats:")

    for row in rows:
        stats = []
        for col in row.find_all('div'):
            stats.append(col.text.strip())
        print(stats)

else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
import pandas as pd


url = "https://bo3.gg/matches/faze-vs-natus-vincere-31-03-2024"

response = requests.get(url)

if response.status_code == 200:
    
    soup = BeautifulSoup(response.content, "html.parser")

    rows = soup.find_all('div', class_='table-row')

    player_stats = []
    for row in rows:
        stats = []
        for col in row.find_all('div'):
            stats.append(col.text.strip())
        player_stats.append(stats)

    df = pd.DataFrame(player_stats)
    df.to_csv('player_stats.csv', index=False)

    print("Player stats have been saved to player_stats.csv.")
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
#CLeaning

df = pd.read_csv('player_stats.csv')
TeamA = df.iloc[0, df.columns.get_loc('0')]
TeamB = df.iloc[7, df.columns.get_loc('0')]
# print(df.head())
df = df.drop(columns=['1','2','3','7'] , errors='ignore')
df = df.drop(index=range(14, 47), errors='ignore')
df = df.drop(index = [0,6,7,13])
df = df.rename(columns={'0': 'Name' ,'4': 'Kills','5': 'Deaths','6': 'Assists','8': 'ADR' })
df['Name'] = df['Name'].str.replace('MVP', '', regex=True)

filtered = df.iloc[:, 0:5]

newcolumn = [TeamA] * 5 + [TeamB] * 5
filtered.insert(0, 'Team', newcolumn)
filtered.reset_index(drop=True, inplace=True)

print(filtered)
```

```python

url = "https://bo3.gg/tournaments/pgl-major-copenhagen-2024/results"

href_links = []

response = requests.get(url)

if response.status_code == 200:
    
    soup = BeautifulSoup(response.content, "html.parser")

    links = soup.find_all('a', class_="c-global-match-link table-cell")
    
    href_links = []
    for link in links:
        href = link.get('href')
        if href:
            href_links.append(href)

    print(href_links)
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")

print(len(href_links))
```

```python
base_url = "https://bo3.gg"

count = 40

for i in href_links:
    
    response = requests.get(f"{base_url}{i}")

    if response.status_code == 200:
        
        soup = BeautifulSoup(response.content, "html.parser")

        rows = soup.find_all('div', class_='table-row')

        player_stats = []
        for row in rows:
            stats = []
            for col in row.find_all('div'):
                stats.append(col.text.strip())
            player_stats.append(stats)

        df = pd.DataFrame(player_stats)
        #print(df.head(15)) # for debugging

        if len(df.columns) > 8:
            df.columns = [str(i) for i in range(len(df.columns))]

        TeamA = df.iloc[0, 0]
        TeamB = df.iloc[7, 0]

      
        df = df.drop(columns=['1', '2', '3', '7'], errors='ignore')
        df = df.drop(index=range(14, 47), errors='ignore')
        df = df.drop(index=[0, 6, 7, 13])
        df = df.rename(columns={'0': 'Name', '4': 'Kills', '5': 'Deaths', '6': 'Assists', '8': 'ADR'}, errors='ignore')

       
        df['Name'] = df['Name'].str.replace('MVP', '', regex=True)

        # filtered = df.iloc[:, 0:5] # Commented out to keep only required columns
        filtered = df[['Name', 'Kills', 'Deaths', 'Assists', 'ADR']]

        newcolumn = [TeamA] * len(df.loc[df.index[:5]]) + [TeamB] * len(df.loc[df.index[5:]])  #list me para hai abhi

        filtered.reset_index(drop=True, inplace=True)
        filtered.insert(0, 'Team', newcolumn)
        filtered = filtered.iloc[:10, :]

        filtered.to_csv(f'2024_Major_{count}.csv', index=False)
        count -= 1

        print(f"Player stats have been saved to 2024_Major_{count}.csv.")
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
```
